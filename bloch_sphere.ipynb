```python
import matplotlib
# matplotlib.use('Qt5Agg')
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
from qutip import *
```

# Pure State $\vert 0 \rangle$

```python
bloch = Bloch()
ket0 = basis(2,0)
bloch.add_states(ket0)
bloch.show()
```

# Pure State $\vert 1 \rangle$

```python
bloch = Bloch()
ket1 = basis(2,1)
bloch.add_states(ket1)
bloch.show()
```

# Qubits in Hadamard Basis

##   $\vert + \rangle = H \vert 0 \rangle$ on the Bloch sphere

```python
H = hadamard_transform() 
print(H*ket0)
bloch = Bloch()
bloch.add_states(H*ket0)
bloch.show()
```

##  The vector  $\vert - \rangle = H \vert 1 \rangle$ on the Bloch sphere

```python
print(H*ket1)
bloch = Bloch()
bloch.add_states(H*ket1)
bloch.show()
```
